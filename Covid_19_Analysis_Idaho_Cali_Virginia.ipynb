Total Covid Hospital Admissions Throughout the Year for each county

```python
import pandas as pd
import calendar

# Create empty lists to store DataFrames
admissions_tables = []
utilization_tables = []

# Assuming 'date_updated' is already in datetime format
for county_name, county_subset in idaho_county_subsets.items():
    county_subset['month'] = county_subset['date_updated'].dt.month_name()
    county_subset['month'] = pd.Categorical(county_subset['month'], categories=calendar.month_name[1:], ordered=True)

    # Group by month and sum to get monthly figures
    monthly_admissions = county_subset.groupby('month')['covid_hospital_admissions'].sum()
    monthly_utilization = county_subset.groupby('month')['covid_inpatient_bed_utilization_figure'].sum()

    # Filter out months with no data
    monthly_admissions = monthly_admissions[monthly_admissions > 0]
    monthly_utilization = monthly_utilization[monthly_utilization > 0]

    # Create a DataFrame for the current county
    admissions_table = pd.DataFrame({
        'County': [county_name],
        'Total Admissions': [monthly_admissions.sum()]
    })

    utilization_table = pd.DataFrame({
        'County': [county_name],
        'Total Utilization': [monthly_utilization.sum()]
    })

    # Append the DataFrame to the lists
    admissions_tables.append(admissions_table)
    utilization_tables.append(utilization_table)

# Concatenate all DataFrames into single tables
total_admissions_table = pd.concat(admissions_tables, ignore_index=True)
total_utilization_table = pd.concat(utilization_tables, ignore_index=True)

# Display the resulting tables
print("Total Covid Hospital Admissions Throughout the Year:")
total_admissions_table


```
